```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_size = [1453, 1000, 700, 350, 175, 100]
```

```python
model_list = [
  "lenet", 
  "mlp4",
  "vgg",
  "lstm1v0",
  "lstm1",
  "lstm2",
  "blstm1",
  "blstm2",
  "lstmfcn",
  "resnet"]
```

```python
data = []
for name in model_list:
    all_df = []
    for ii in range(6):
        results = pd.read_csv(f'{name}_aug_{ii + 1}_eval.csv')
        results['train_size'] = train_size[ii]
        all_df.append(results)
    df = pd.concat(all_df)
    data.append(df)
```

```python
data[0][data[0]["Metric"] == "mad"].groupby("train_size").mean()["Value"]
```

```python
fig, ax = plt.subplots()
ax.plot(train_size[::-1], np.vstack(data[i][data[i]["Metric"] == "mae"].groupby("train_size").mean()["Value"]
                          for i in range(10)).T, 'o--', label=model_list)
ax.set_ylim([1.5,3])
ax.set_xlim([0,2000])
ax.legend()
ax.set_title('Mean absolute error with augmentation')
```

```python
fig, ax = plt.subplots()
ax.plot(train_size[::-1], np.vstack(data[i][data[i]["Metric"] == "r2"].groupby("train_size").mean()["Value"]
                          for i in range(10)).T, 'o--', label=model_list)
ax.set_ylim([0,1])
ax.set_xlim([0,2000])
ax.legend()
ax.set_title('R2 with augmentation')
```

```python
fig, ax = plt.subplots()
ax.plot(train_size[::-1], np.vstack(data[i][data[i]["Metric"] == "mad"].groupby("train_size").mean()["Value"]
                          for i in range(10)).T, 'o--', label=model_list)
ax.set_ylim([1.2,2.7])
ax.set_xlim([0,2000])
ax.legend()
ax.set_title('Median absolute deviation with augmentation')
```
